```python
def findDescriptors(image, method):
    if method == "sift":
        descriptor = cv2.xfeatures2d.SIFT_create()
    elif method == "surf":
        descriptor = cv2.xfeatures2d.SURF_create()
    
    (kps, features) = descriptor.detectAndCompute(image,None)
    return (kps, features)
```

```python
kpsA, featuresA = findDescriptors(image1_gray, method=feature_extractor)
kpsB, featuresB = findDescriptors(image2_gray, method=feature_extractor)
```

```python
plt.imshow(cv2.drawKeypoints(image1_gray,kpsA,None,color=(0,255,0)))
```

```python
def createMatcher(method,crossCheck):
    "Create and return a Matcher Object"
    
    if method == 'sift' or method == 'surf':
        bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=crossCheck)
    elif method == 'orb' or method == 'brisk':
        bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=crossCheck)
    return bf
```
